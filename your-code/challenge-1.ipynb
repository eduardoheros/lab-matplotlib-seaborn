#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
x=np.arange(0,100)
y1=x**2
y2=np.exp(x)
plt.subplots(facecolor="#fffee8",figsize=(6,6))
plt.plot(x,y1,color="g", lw=2, ls=':')
plt.xlabel("Pruebas realizadas")
plt.ylabel("Personas detectadas")
plt.plot(x,y2,color="orange", lw=2, ls='-')
plt.xlabel("Horas")
plt.ylabel("Carga viral")
plt.xscale("log")
plt.yscale("log")
plt.show();

```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
vehicles=pd.read_csv("https://raw.githubusercontent.com/eduardoheros/lab-matplotlib-seaborn/master/your-code/vehicles.csv")
x=vehicles["Combined MPG"]
y=vehicles["Highway MPG"]
```

```python
plt.figure(figsize=(12,6),facecolor="white")
plt.scatter(x,y,alpha=0.5)
plt.xlabel(x.name)
plt.ylabel(y.name);
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
plt.figure(figsize=(10,10),facecolor="#47784a")
plt.boxplot(vehicles['CO2 Emission Grams/Mile'],showfliers=False);
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
plt.figure(facecolor="#d1e6ff",figsize=(10,7))
plt.hist(vehicles['Fuel Barrels/Year'],bins=20,density=True,rwidth=0.95)
plt.xlabel("Fuel Barrels/Year")
plt.ylabel("Probability")
plt.title("Histogram");
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
fueltype_citympg=vehicles.groupby("Fuel Type")["City MPG"].mean()
fueltype_citympg.sort_values(axis=0,ascending=True,inplace=True)
arange_xticks=np.arange(len(fueltype_citympg))
```

```python
colors=["#f94144","#f3722c","#f8961e","#f9c74f","#90BE6D","#43AA8B","#577590","#f94144","#f3722c","#f8961e","#f9c74f","#90BE6D","#43AA8B"]
plt.figure(figsize=(10,5),facecolor="white")
plt.barh(arange_xticks,fueltype_citympg.values,color=colors)
plt.yticks(arange_xticks,fueltype_citympg.index);
```
